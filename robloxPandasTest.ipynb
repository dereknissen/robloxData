```python
# Import dependenceis
import pandas as pd
import requests
```

```python
# Cleaning
robloxData = pd.read_json('robloxData.json')
robloxData = robloxData.transpose() # Flip columns and rows
robloxData.head()
```

```python
# Import usernames
url = 'https://users.roblox.com/v1/users/'
usernames = []
for row in robloxData.index:
    username = "n/a"
    try:
        response = requests.get(url + str(row)).json()
        username = response["name"]
    except:
        pass
    usernames.append(username)

robloxData["username"] = usernames
robloxData["userId"] = robloxData.index
robloxData = robloxData.set_index("username")
robloxData.head()
```

```python
# Sort by money
topMoney = robloxData.copy()
topMoney = pd.DataFrame(topMoney["money"])
topMoney = topMoney.sort_values("money", ascending = False)
topMoney.head(10)
```
